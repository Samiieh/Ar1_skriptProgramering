```python
# Made a variable called poke, we are going to remove Null data  
poke = data.dropna()
poke
```

```python
# The Data below is showing pokemon which the Total is over 500 and below 1000, and Attack less then 100 and speed greater than 100.
poke[((poke['Total'] >= 500) & (poke['Total'] < 1000))
    & (poke['Attack'] < 100)
    & (poke['Speed'] > 100)]
```

```python
# Sorting Data values by HP, and grouping by # and Name, printing the top 15 rows

poke.groupby(['#','Name'])[['HP']].mean().sort_values(['HP'], ascending=False).iloc[0:15]
```

```python
# Checking the Null-values in rows
poke.isnull().sum()
```

```python
# Dropping all dtypes exept int64 and making a new variable called dropping, for future use in heat map

dropping = poke.drop(['Legendary','Type 1','Type 2','Name'], axis=1)
```

```python
# Printing information about Data
dropping.info()
```

```python
# Information about dataset, 414 rows and 9 columns
dropping.shape
```

```python
# Creating 2 variables, one to store the mean of the HP and second one to print and format the result with 1 decimal digit.

hp_mean = dropping['HP'].mean()
sentence = "The mean HP of all Pokémon is: {:.1f}".format(hp_mean)
print(sentence)
```

```python
# Create a function to rate Pokémons based on attack value

def stats_group(attack):
    if attack >= 155:
        return 'Exellent'
    elif attack >= 80:
        return 'Average'
    else:
        return 'Bad'
```

```python

#Applying the rating function to our poke data, creating a new variable in the dataset to hold the Attack_rating category

poke['Attack_rating'] = poke['Attack'].apply(stats_group)
```
